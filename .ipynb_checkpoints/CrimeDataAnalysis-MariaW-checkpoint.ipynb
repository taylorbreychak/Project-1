```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.stats as stats
import numpy as np
```

```python
Zillow_Ohio_csv = "Resources/newZillowOH_data.csv"
# crime2017_csv = "Resources/OhioCrime2017_NoNan.csv"
# crime2016_csv = "Resources/OhioCrime2016_NoNan.csv"
Ohio_Crime_csv = "Resources/CrimeData_merged.csv"
```

```python
zillow_df = pd.read_csv(Zillow_Ohio_csv)
zillow_df = zillow_df.fillna(0)
ohiocrime_df = pd.read_csv(Ohio_Crime_csv)
# #crime2017_df.dropna(how='all')

# crime2017_df = crime2017_df.fillna(0)

# crime2017_df['County'] = crime2017_df['County'].str.upper()
# crime2017_df.head()
# #crime2017_df.columns

# crime2016_df = pd.read_csv(crime2016_csv)
# #crime2017_df.dropna(how='all')

# crime2016_df = crime2016_df.fillna(0)

# crime2016_df['County'] = crime2016_df['County'].str.upper()
# crime2016_df.head()
```

```python
#zillow_df['CountyName'] = zillow_df['CountyName'].str.upper()
#zillow_df.head()
zillow_df = zillow_df.rename(columns={"RegionName": "City", "2017": "2017 Median Price", "2016":"2016 Median Price",
                                     "2015": "2015 Median Price", "2014": "2014 Median Price", "2013":"2013 Median Price",
                                     "2012":"2012 Median Price"})
zillow_df.columns
```

```python
# merge_Crime_df = pd.merge(crime2016_df, crime2017_df, left_on="City", right_on="City")
# merge_Crime_df.head()
# merge_Crime_df.to_csv("merge_Crime_df1.csv")

# merge_Crime_df.columns
```

```python
merge_Zillow_Crime_df = pd.merge(ohiocrime_df, zillow_df, on = "City")
#merge_Zillow_Crime_df = merge_Zillow_Crime_df.fillna(0)
merge_Zillow_Crime_df.to_csv("MergeDataMaria.csv")
merge_Zillow_Crime_df.head()

#merge_Zillow_Crime_df.columns

merge_Zillow_Crime_df=merge_Zillow_Crime_df.sort_values("2017 Median Price", ascending=False)
merge_Zillow_Crime_df.reset_index()
merge_Zillow_Crime_df
```

```python
#Tumor_Response_semUpdate = Tumor_Response_sem.pivot(index="Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

merge_Zillow_Crime_df
merge_Zillow_Crime_df["Crime Rate 2017"] = ((merge_Zillow_Crime_df["2017 Violent Crime"])/(merge_Zillow_Crime_df["2017 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2016"] = ((merge_Zillow_Crime_df["2016 Violent Crime"])/(merge_Zillow_Crime_df["2016 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2015"] = ((merge_Zillow_Crime_df["2015 Violent Crime"])/(merge_Zillow_Crime_df["2015 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2014"] = ((merge_Zillow_Crime_df["2014 Violent Crime"])/(merge_Zillow_Crime_df["2014 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2013"] = ((merge_Zillow_Crime_df["2013 Violent Crime"])/(merge_Zillow_Crime_df["2013 Population"]))*100000
#merge_Zillow_Crime_df["Crime Rate 2012"] = ((merge_Zillow_Crime_df["2012 Violent Crime"])/(merge_Zillow_Crime_df["2012 Population"]))*100000
merge_Zillow_Crime_df.reset_index().tail()
```

```python
Crime_PropertyValue_2017df = merge_Zillow_Crime_df[["City","2017 Median Price", "Crime Rate 2017", "2017 Population"]]
Crime_PropertyValue_2017df.head()
# citydata=Crime_PropertyValue_2017df.loc["City"]== "New Albany", ["Median Price", "2017 Population"]
```

```python
# plt.scatter(Crime_PropertyValue_2017df["Crime Rate 2017"], Crime_PropertyValue_2017df["2017 Median Price"], color="teal", marker="o", edgecolor="black")

# plt.title("2017 Housing Median Sale vs. Violent Crime Rate")
# plt.ylabel("Housing Median Sale")
# plt.xlabel("Violent Crime Rate (per 100,000)")
# plt.grid(True)
# plt.show()
# plt.savefig("Plots/2017 Housing Median Sale vs. Violent Crime Rate.png")


crime_rate= Crime_PropertyValue_2017df["Crime Rate 2017"]
median_housing = Crime_PropertyValue_2017df["2017 Median Price"]
population = Crime_PropertyValue_2017df["2017 Population"]
city = Crime_PropertyValue_2017df["City"]
fig, ax = plt.subplots()
fig.suptitle("2017 Ohio Crime Rate vs Median Housing Sale ", fontsize=16, fontweight="bold")
# fig.suptitle("2017 Median Housing Sale vs. Violent Crime Rate", fontsize=16, fontweight="bold"
scatter = ax.scatter(crime_rate, median_housing, c=population, s=population/1000, cmap="Spectral")
                
legend1 = ax.legend(*scatter.legend_elements(),
                     loc="upper center", title="City")
ax.add_artist(legend1)
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.1)
legend2 = ax.legend(handles, labels, loc="upper right", title="Population")
#plt.title("2017 Median Rental Value vs. Violent Crime Rate")
plt.ylabel("Median Rental Value")
plt.xlabel("Violent Crime Rate")
plt.grid(True)
plt.show()
```

```python
Data1=pd.to_numeric(Crime_PropertyValue_2017df["Crime Rate 2017"])
Data2=pd.to_numeric(Crime_PropertyValue_2017df["2017 Median Price"])
```

```python
# Set line
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(Data1 , Data2)
fit = vc_slope * Data1 + vc_int
# Plot data
fig, ax = plt.subplots()
fig.suptitle("2017 Ohio Crime Rate vs Median Housing Sale ", fontsize=16, fontweight="bold")
#ax.set_xlim(0, 10)
#ax.set_ylim(0, 10)
plt.grid()
ax.set_xlabel("Crime Rate (Per 100,000)")
ax.plot(Data1, Data2, linewidth=0, marker='o',color="teal")
ax.plot(Data1, fit, "b--", linewidth=1)
ax.set_ylabel("Median Housing Sale")
print(vc_p)
plt.title("Pvalue ="+str(4.855181777717296e-07))
plt.show()
```

```python
#Ohio Avg. Crime Rate

CrimeRate_df = merge_Zillow_Crime_df[["City","Crime Rate 2013", "Crime Rate 2014",
                                      "Crime Rate 2015", "Crime Rate 2016","Crime Rate 2017"]]
CrimeRate_df = CrimeRate_df.rename(columns={"Crime Rate 2016": "2016", "Crime Rate 2015": "2015",
                                            "Crime Rate 2014": "2014",  "Crime Rate 2013": "2013",
                                           "Crime Rate 2017": "2017"})

CrimeRate_df =CrimeRate_df.fillna(0)
CrimeRate_df= CrimeRate_df.set_index("City")
AverageOH_Crime = CrimeRate_df.mean()
AverageOH_Crime
CrimeRate_df.head(15)
```

```python
figsize=(10,8)
fig, ax1 = plt.subplots(figsize=(10,8))

years = CrimeRate_df.keys()
OhioAvg, = plt.plot(years, AverageOH_Crime, color="black", marker="*", label="Ohio Average Crime" )

City1, = plt.plot(years, CrimeRate_df.loc['Cleveland',["2013","2014","2015","2016","2017"]], 
                          color="green", marker = "o", label="Cleveland")
City2, = plt.plot(years, CrimeRate_df.loc['Columbus',["2013","2014","2015","2016","2017"]], 
                          color="red", marker = "o",label="Columbus")
City3 = plt.plot(years, CrimeRate_df.loc['Toledo',["2013","2014","2015","2016","2017"]], 
                          color="blue",  marker = "o", label="Toledo")
City4 = plt.plot(years, CrimeRate_df.loc['Akron',["2013","2014","2015","2016","2017"]], 
                          color="purple", marker = "o",label="Akron")
City5 = plt.plot(years, CrimeRate_df.loc['Cincinnati',["2013","2014","2015","2016","2017"]], 
                         color="yellow", marker = "o",label="Cincinnati")
City6 = plt.plot(years, CrimeRate_df.loc['Parma',["2013","2014","2015","2016","2017"]], 
                         color="teal", marker = "o",label="Parma")
City7 = plt.plot(years, CrimeRate_df.loc["Canton",["2013","2014","2015","2016","2017"]], 
                         color="orange",  marker = "o", label="Canton")
City8 = plt.plot(years, CrimeRate_df.loc["Lorain",["2013","2014","2015","2016","2017"]], 
                         color="magenta", marker = "o",label="Lorain")
City9 = plt.plot(years, CrimeRate_df.loc["West Chester",["2013","2014","2015","2016","2017"]], 
                          color="coral", marker = "o", label= "West Chester")
City10 = plt.plot(years, CrimeRate_df.loc["Hamilton",["2013","2014","2015","2016","2017"]], 
                          color="olive", marker = "o",label="Hamilton")

plt.title("Average Ohio Crime vs Top 10 City by Population")
plt.ylabel("Violent Crime Rate")
plt.xlabel("Year")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
