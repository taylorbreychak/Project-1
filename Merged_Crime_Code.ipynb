```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
OHcrime_2017 = "Resources/OhioCrime2017_NoNan.csv"
OHcrime_2016 = "Resources/OhioCrime2016_NoNan.csv"
OHcrime_2015 = "Resources/OhioCrime2015_NoNan.csv"
OHcrime_2014 = "Resources/OhioCrime2014_NoNan.csv"
OHcrime_2013 = "Resources/OhioCrime2013_NoNan.csv"
OHcrime_2012 = "Resources/OhioCrime2012_NoNan.csv"

OHcrime_2017_df = pd.read_csv(OHcrime_2017)
OHcrime_2016_df = pd.read_csv(OHcrime_2016)
OHcrime_2015_df = pd.read_csv(OHcrime_2015)
OHcrime_2014_df = pd.read_csv(OHcrime_2014)
OHcrime_2013_df = pd.read_csv(OHcrime_2013)
OHcrime_2012_df = pd.read_csv(OHcrime_2012)
OHcrime_2015_df
```

```python
combined_crime_df = pd.merge(OHcrime_2017_df, OHcrime_2016_df, how = 'left', on="City")

#combined_crime_df.reset_index(drop=True)
#combined_crime_df.to_csv("Resources/CrimeData_merged.csv")
combined_crime_df
```

```python
combined_crime_df = combined_crime_df.rename(columns={"POPULATION_x":"2017 Population",
                                                              "VIOLENT CRIME_x":"2017 Violent Crime",
                                                              "MURDER_x":"2017 Murder", "RAPE_x": "2017 Rape",
                                                              "ROBBERY_x": "2017 Robbery", "AGGRAVATED ASSAULT_x": 
                                                              "2017 Aggravated Assault"})

# Rename our _y columns to 2016
combined_crime_df = combined_crime_df.rename(columns={"POPULATION_y":"2016 Population",
                                                              "VIOLENT CRIME_y":"2016 Violent Crime",
                                                              "MURDER_y":"2016 Murder", "RAPE_y": "2016 Rape",
                                                              "ROBBERY_y": "2016 Robbery", "AGGRAVATED ASSAULT_y": 
                                                              "2016 Aggravated Assault"})
combined_crime_df.head()
```

```python
# Merge our newly combined dataframe with the 2015 dataframe

combined_crime_df = pd.merge(combined_crime_df, OHcrime_2015_df, how = 'outer', on="City")
combined_crime_df = combined_crime_df.rename(columns={"POPULATION":"2015 Population",
                                                        "VIOLENT CRIME":"2015 Violent Crime",
                                                        "MURDER":"2015 Murder", "RAPE": "2015 Rape",
                                                        "ROBBERY": "2015 Robbery", "AGGRAVATED ASSAULT": 
                                                        "2015 Aggravated Assault"})

combined_crime_df = pd.merge(combined_crime_df, OHcrime_2014_df, how = 'outer', on="City")
combined_crime_df = combined_crime_df.rename(columns={"POPULATION":"2014 Population",
                                                      "VIOLENT CRIME":"2014 Violent Crime",
                                                      "MURDER":"2014 Murder", "RAPE": "2014 Rape",
                                                      "ROBBERY": "2014 Robbery", "AGGRAVATED ASSAULT": 
                                                    "2014 Aggravated Assault"})
combined_crime_df

combined_crime_df.to_csv("Resources/CrimeData_merged.csv")
# #combined_crime_df.columns
```

```python
violent_crimetable = combined_crime_df[['City','County_x','2017 Population', '2017 Violent Crime',
         '2016 Population', '2016 Violent Crime', '2015 Population', '2015 Violent Crime', '2014 Population', '2014 Violent Crime']]

violent_crimetableGrouped = violent_crimetable.groupby("City")
Final_CrimeRateData= violent_crimetableGrouped.sum()
Final_CrimeRateData
#new_df = pd.DataFrame(violent_crimetableGrouped["2017 Population", "2017 Violent Crime", "2016 Population", "2016 Violent
                      #Crime"].sum()

#DataFrame without City duplicates:
Final_CrimeRateData
```

```python

Final_CrimeRateData["Crime Rate 2017"] = ((Final_CrimeRateData["2017 Violent Crime"])/(Final_CrimeRateData["2017 Population"]))*100
Final_CrimeRateData["Crime Rate 2016"] = ((Final_CrimeRateData["2016 Violent Crime"])/(Final_CrimeRateData["2016 Population"]))*100
Final_CrimeRateData["Crime Rate 2015"] = ((Final_CrimeRateData["2015 Violent Crime"])/(Final_CrimeRateData["2015 Population"]))*100
Final_CrimeRateData["Crime Rate 2014"] = ((Final_CrimeRateData["2014 Violent Crime"])/(Final_CrimeRateData["2014 Population"]))*100

Final_CrimeRateData.head()
# #violent_crimetableUpdate = violent_crimetable.sort_values("2017 Population", ascending=False)
Final_CrimeRateData.to_csv("Resources/CrimeRate.csv")
```
