```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Zillow_Ohio_csv = "Resources/newZillowOH_data.csv"
# crime2017_csv = "Resources/OhioCrime2017_NoNan.csv"
# crime2016_csv = "Resources/OhioCrime2016_NoNan.csv"
Ohio_Crime_csv = "Resources/CrimeData_merged.csv"
```

```python
zillow_df = pd.read_csv(Zillow_Ohio_csv)
zillow_df = zillow_df.fillna(0)
ohiocrime_df = pd.read_csv(Ohio_Crime_csv)
# #crime2017_df.dropna(how='all')

# crime2017_df = crime2017_df.fillna(0)

# crime2017_df['County'] = crime2017_df['County'].str.upper()
# crime2017_df.head()
# #crime2017_df.columns

# crime2016_df = pd.read_csv(crime2016_csv)
# #crime2017_df.dropna(how='all')

# crime2016_df = crime2016_df.fillna(0)

# crime2016_df['County'] = crime2016_df['County'].str.upper()
# crime2016_df.head()
```

```python
#zillow_df['CountyName'] = zillow_df['CountyName'].str.upper()
#zillow_df.head()
zillow_df = zillow_df.rename(columns={"RegionName": "City", "2017": "2017 Median Price", "2016":"2016 Median Price",
                                     "2015": "2015 Median Price", "2014": "2014 Median Price"})
zillow_df.columns
```

```python
# merge_Crime_df = pd.merge(crime2016_df, crime2017_df, left_on="City", right_on="City")
# merge_Crime_df.head()
# merge_Crime_df.to_csv("merge_Crime_df1.csv")

# merge_Crime_df.columns
```

```python
merge_Zillow_Crime_df = pd.merge(ohiocrime_df, zillow_df, how='left', left_on="City", right_on="City")

merge_Zillow_Crime_df.to_csv("MergeDataMaria.csv")
merge_Zillow_Crime_df.head()
#merge_Zillow_Crime_df.columns

merge_Zillow_Crime_df=merge_Zillow_Crime_df.sort_values("2017 Population", ascending=False)
merge_Zillow_Crime_df
```

```python
#Tumor_Response_semUpdate = Tumor_Response_sem.pivot(index="Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

merge_Zillow_Crime_df
merge_Zillow_Crime_df["Crime Rate 2017"] = ((merge_Zillow_Crime_df["2017 Violent Crime"])/(merge_Zillow_Crime_df["2017 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2016"] = ((merge_Zillow_Crime_df["2016 Violent Crime"])/(merge_Zillow_Crime_df["2016 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2015"] = ((merge_Zillow_Crime_df["2015 Violent Crime"])/(merge_Zillow_Crime_df["2015 Population"]))*100000
merge_Zillow_Crime_df["Crime Rate 2014"] = ((merge_Zillow_Crime_df["2014 Violent Crime"])/(merge_Zillow_Crime_df["2014 Population"]))*100000
merge_Zillow_Crime_df.head()
```

```python
Crime_PropertyValue_2017df = merge_Zillow_Crime_df[["City","2017 Median Price", "Crime Rate 2017"]]
Crime_PropertyValue_2017df.head()
```

```python
plt.scatter(Crime_PropertyValue_2017df["Crime Rate 2017"], Crime_PropertyValue_2017df["2017 Median Price"], marker="o", edgecolor="black")

plt.title("2017 Median Rental Value vs. Violent Crime Rate")
plt.ylabel("Median Rental Value")
plt.xlabel("Violent Crime Rate")
plt.grid(True)
plt.show()
```
